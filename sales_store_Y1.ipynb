```python
import pandas as pd
import matplotlib as plt
import seaborn as sns  
```

```python
df0 = pd.read_csv('sales_store.csv')
df0.head()
```

```python
df0.info()
```

```python
df1 = df0.copy()
df1['Date'] = pd.to_datetime(df1["Date"], format="%d/%m/%Y")
df1.info()
```

```python
df2 = df1.copy() 
df2['IsHoliday'] = df2['IsHoliday'].replace(
    {
        True: 1,
        False: 0
    }
)
df2.head()
```

```python
df2['year'] = pd.DatetimeIndex(df2['Date']).year
df2.head()
```

```python
df_2010 = df2[df2['year'] == 2010]
df_2010.reset_index(inplace=True)
df_2010.head()
```

```python
df_2011 = df2[df2['year'] == 2011]
df_2011.reset_index(inplace=True)
df_2011.info()
```

```python
df_2012 = df2[df2['year'] == 2012]
df_2012.reset_index(inplace=True)
df_2012.info()
```

```python
#exportar a csv
df_2010.to_csv('data_2010.csv')
df_2011.to_csv('data_2011.csv')
df_2012.to_csv('data_2012.csv')
```
